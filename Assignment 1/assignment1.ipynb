```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('used_cars_data.csv')
dataset
```

```python
from datetime import date
date.today().year
dataset['Car_Age']=date.today().year-dataset['Year']
dataset.head()
```

```python
dataset['Brand'] = dataset.Name.str.split().str.get(0)
dataset['Model'] = dataset.Name.str.split().str.get(1) + dataset.Name.str.split().str.get(2)
dataset[['Name','Brand','Model']]
```

```python
print(dataset.Brand.unique())
print(dataset.Brand.nunique())
```

```python
searchfor = ['Isuzu' ,'ISUZU','Mini','Land']
dataset[dataset.Brand.str.contains('|'.join(searchfor))].head(5)
```

```python
dataset["Brand"].replace({"ISUZU": "Isuzu", "Mini": "Mini Cooper","Land":"Land Rover"}, inplace=True)
```

```python
dataset.describe().T
```

```python
dataset.describe(include='all').T
```

```python
cat_cols=dataset.select_dtypes(include=['object']).columns
num_cols = dataset.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(dataset[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    dataset[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=dataset[col])
    plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize = (18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')
sns.countplot(ax = axes[0, 0], x = 'Fuel_Type', data = dataset, color = 'blue', 
              order = dataset['Fuel_Type'].value_counts().index);
sns.countplot(ax = axes[0, 1], x = 'Transmission', data = dataset, color = 'blue', 
              order = dataset['Transmission'].value_counts().index);
sns.countplot(ax = axes[1, 0], x = 'Owner_Type', data = dataset, color = 'blue', 
              order = dataset['Owner_Type'].value_counts().index);
sns.countplot(ax = axes[1, 1], x = 'Location', data = dataset, color = 'blue', 
              order = dataset['Location'].value_counts().index);
sns.countplot(ax = axes[2, 0], x = 'Brand', data = dataset, color = 'blue', 
              order = dataset['Brand'].head(20).value_counts().index);
sns.countplot(ax = axes[2, 1], x = 'Model', data = dataset, color = 'blue', 
              order = dataset['Model'].head(20).value_counts().index);
axes[1][1].tick_params(labelrotation=45);
axes[2][0].tick_params(labelrotation=90);
axes[2][1].tick_params(labelrotation=90);
```
